### 1. Generic Exception Handling


```python
help(Exception)
```

```python
def division2():
    try:
        num = float(input("Please enter numerator :"))
        den = float(input("Please enter denominator :"))
        return num / den
    except Exception as e :
        print(f"Exception occured : {e}")
```

```python
# num = 45, den = 12
d5 = division2()
print(d5)
print("Program Completed")
```

```python
# num = 17, den = 0
d6 = division2()
print(d6)
print("Program Completed")
```

```python
# num = 14, den = 0
d7 = division2()
print(d7)
print("Program Completed")
```

```python
# Where we know operation can fail we can use try and except block
```

```python
# num = "b"
d8 = division2()
print(d8)
print("Program Completed")
```

```python
# num = 23, den = blank just press enter
d9 = division2()
print(d9)
print("Program Completed")
```
