```python
# We can create a normal class or dataclass here...
from dataclasses import dataclass
```

```python
@dataclass
class Triangle:  # We have to provide 3 sides of the triangle
    a: int | float
    b: int | float
    c: int | float

    def __post_init(self):
        errors = []

        if not isinstance(self.a, (int|float)) or self.a <= 0:
            errors.append("Side a should be a positive integer or float")
        if not isinstance(self.b, (int|float)) or self.b <= 0:
            errors.append("Side b should be a positive integer or float")
        if not isinstance(self.c, (int|float)) or self.c <= 0:
            errors.append("Side c should be a positive integer or float")
        if not self.is_triangle():
            errors.append("Sum of any two sides should always be greater than third")
        
        if errors:
            raise ValueError("\n".join(errors))

    def is_triangle(self): # Multiple conditions can be written in round bracket likewise
        if (
            self.a + self.b > self.c and
            self.b + self.a > self.a and
            self.c + self.a > self.b
        ):
            return True 
        else:
            return False # If conditions are validated it forms a triangle if not condition becomes false.
        
    def perimeter(self):
        return self.a + self.b + self.c
    
    def area(self):
        s = self.perimeter() / 2
        a = (s * (s - self.a) * (s - self.b)*(s - self.c))**(1/2)
        return a
```

### Calculating Area and perimeter for different shapes

```python
@dataclass
class Rectangle:
    width: int|float
    height: int|float

    def perimeter(self):
        return 2 * (self.width + self.height)
    
    def area(self):
        return self.width * self.height
```

```python
from math import pi
```

```python
pi
```

```python
@dataclass
class Circle:
    radius: int|float

    def perimeter(self):
        return 2 * pi * self.radius
    
    def area(self):
        return pi * (self.radius ** 2)
```

```python
s1 = Triangle(12,13,14)
s2 = Rectangle(20,10)
s3 = Circle(14)
s4 = Triangle(3,4,5)
s5 = Circle(21)
s6 = Rectangle(5,3)
```

```python
shapes = [s1,s2,s3,s4,s5,s6]
```

```python
for i in shapes:
    print(i)
    print(type(i))
    print(f"perimeter : {i.perimeter():.4f}")
    print(f"Area: {i.area():.4f}")
    print("\n===========================\n")
```
