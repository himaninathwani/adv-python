### Multithreading allows working on multiple tasks simultaneously

```python
import time
```

```python
%%time 
time.sleep(2)
print("Example")
time.sleep(3)
print("Hello")
```

```python
# this is magic command to print time as well %% time
```

```python
def simple_interest(p,n,r):
    print("Simple interest execution started")
    time.sleep(3)
    i = (p * n * r)/100
    a = p + i
    print(f"Interest : {i:.2f} INR")
    print(f"Amount : {a:.2f} INR")
    print("Simple Interest Execution Completed")
```

```python
%%time
simple_interest(p=57000, n=5, r=7.2)
```

```python
def hypotenuse(a, b):
    print("Hypotenuse function execution started")
    time.sleep(2)
    c = (a**2 + b**2) ** (1/2)
    print(f"hypotenuse for {a} and {b} is {c:.4f}")
    print("Hypotenuse Execution Completed")
```

```python
%%time
hypotenuse(4, 7)
```

```python
%%time
simple_interest(p=70000,n=3,r=6.5)
hypotenuse(3,4)
```

```python
# Thats how multithreading works
```

### If you want to execute 2 functions simultaneously

```python
from threading import Thread
```

```python
%%time
# Create and start the threads (args = should be tuple value and in sequence only)
th1 = Thread(target=simple_interest, args=(45000, 2, 6.2))
th2 = Thread(target=hypotenuse, args=(5, 9))

# Start all the threads
th1.start()
th2.start()

# Wait for all threads to finish
th1.join()
th2.join()
```

```python
# this is multithreading on multiple function
```

### Single function but multiple values to calculate

```python
def square(n):
    time.sleep(1)
    s = n**2
    print(f"Square of number {n} is {s}\n\n")
```

```python
%%time
square(12.5)
```

```python
a = [1,11,13,5,12,14.5]
```

```python
len(a)
```

```python
%%time
for i in a:
    square(i)
```

```python
# We will try with numbers, more the numbers more time it will take
```

### Multithreading

```python
def square_mul(b: list):
    # Create the threads 
    threads = []
    for i in b:
        th = Thread(target=square, args=(i,))
        th.start()
        threads.append(th)
    
    # Wait for all threads to finish
    for th in threads:
        th.join()
```

```python
a
```

```python
%%time
square_mul(a)
```

```python
b = np.arange(1,101,1)
b
```

```python
%%time
square_mul(b)
```
